```python
# 先对 tag 进行排序，让相同组合必然组成完全一致的 tuple 类型
for member in members:
    member['tags'].append(member['type'] + '干员')
    member['tags'].sort()
```

```python
# 首先跳过不影响 9小时 招募的1/2星与无 tag 干员，以及平时不可能出现的 6 星干员，还有公开招募不出现的干员
# 然后分别遍历保存能出3星以上的 tag 组合，排除资深和高级资深
# 最后在组合中排除掉所有含有3星的，就得到必出4星的。5星类推
opened3 = {} # 因为包含被排除的 tag 组合

import itertools
# 取最多3个 tag 的组合结果
def combine3(l):
    re = []
    re.extend(itertools.combinations(l,1))
    re.extend(itertools.combinations(l,2))
    re.extend(itertools.combinations(l,3))
    return re
for member in members:
    level = member['level']
    tags = member['tags']
    hidden = member['hidden']
    if hidden or level < 3 or level == 6 or len(''.join(tags)) == 0:
        continue
    name = member['name']
    allKeys = combine3(tags)
    for key0 in allKeys:
        if '资深干员' in key0 or '高级资深干员' in key0:
            continue
        key = tuple(key0)
        if key not in opened3:
            opened3[key] = []
        opened3[key].append(member)
```

```python
# 筛选只出现 4/5 星的组合
opened4 = {}
opened5 = {}
for (k,v) in opened3.items():
    levels = tuple(map(lambda i:i['level'], v))
    if 3 not in levels:
        opened4[k] = v
        if 4 not in levels:
            opened5[k] = v
```

```python
# 从 4星 组合中排除掉纯 5星 组合，减少重复
for k in opened5.keys():
    if k in opened4:
        del opened4[k]
```

```python
# 排除冗余 tag 即多选 tag 后不影响出现的人物
def excludeDup(opened0, ignoreMember=False):
    opened = {}
    for (k,v) in opened0.items():
        opened[k] = tuple(map(lambda i:i['name'], v))
    closed = {}
    for (k,v) in opened.items():
        allKeys = combine3(k)
        for key in allKeys:
            if k == key:
                continue
            l = list(key)
            if not ignoreMember:
                l.append(v)
            l = tuple(l)
            if l not in closed:
                closed[l] = []
            closed[l].append(k)
    for (k,v) in opened.items():
        key = list(k)
        if not ignoreMember:
            key.append(v)
        key = tuple(key)
        if key in closed:
            for ok in closed[key]:
                if ok in opened0:
                    del opened0[ok]

excludeDup(opened4)
excludeDup(opened5)
```

```python
# 输出结果
def printTag(k, v):
    print(','.join(k), '\t', ','.join(list(map( lambda s:s['name'], v))))
print('4星tag总数',len(opened4.keys()))
for (k,v) in opened4.items():
    printTag(k, v)

print('-------------')
print('5星tag总数',len(opened5.keys()))
for (k,v) in opened5.items():
    printTag(k, v)
```

```python
# 在上一步基础上，直接对同星级的组合内部，排除星级冗余 tag 即多选 tag 后不影响出现的星级
excludeDup(opened4, True)
excludeDup(opened5, True)
```

```python
# 再做一点融合、分列
# 分列按照位置、职业、词缀1、词缀2四列
openedl = []
poss = ['近战位', '远程位']
pros = ['医疗干员',
'先锋干员',
'重装干员',
'辅助干员',
'特种干员',
'狙击干员',
'术师干员',
'近卫干员'
]
for (k,v) in opened4.items():
    openedl.append({'tags':k, 'pos':'', 'pro':'', 'dis1':'', 'dis2':'', 'result':v, 'minLevel':4})
for (k,v) in opened5.items():
    openedl.append({'tags':k, 'pos':'', 'pro':'', 'dis1':'', 'dis2':'', 'result':v, 'minLevel':5})

for item in openedl:
    tags = item['tags']
    for tag in tags:
        if tag in poss:
            item['pos'] = tag
        elif tag in pros:
            item['pro'] = tag
        elif item['dis1'] == '':
            item['dis1'] = tag
        else:
            item['dis2'] = tag
```

```python
# 再对主要的4星词缀进行简化，得出最需要记住的基础词缀，把通过添加词缀可以升级为纯5星的组合往后排（不那么重要，因为首先要有基础词缀）
openedAD = {} # 将四五星组合融合起来映射
for item in openedl:
    tags = item['tags']
    item['canUpdate'] = 0 # 默认不可升级
    item['rank'] = 1 # 重置默认优先级最高
    openedAD[tags] = item

for item in openedl:
    if item['minLevel'] == 4:
        continue
    tags = item['tags']
    allKeys = combine3(tags)
    for key in allKeys:
        # 存在包含关系时，被包含者为4星时设置为可升级，包含者这种为更低优先级
        if key != tags and key in openedAD:
            if openedAD[key]['minLevel'] == 4:
                openedAD[key]['canUpdate'] = 1
            openedAD[key]['rank'] = 2
            openedAD[tags]['rank'] = 3

# 将容易记的单tag标记为最低优先级
for item in openedl:
    if len(item['tags']) == 1:
        item['rank'] = 4
```

```python
# 最后按特定排序，并输出
openedl.sort(key=lambda i:(i['dis2']), reverse=True)
openedl.sort(key=lambda i:(i['dis1']), reverse=True)
openedl.sort(key=lambda i:(i['pro']), reverse=True)
openedl.sort(key=lambda i:(i['pos']), reverse=True)
openedl.sort(key=lambda i:(i['canUpdate']), reverse=False)
openedl.sort(key=lambda i:(i['rank']), reverse=False)
sep = '\t'
defaultValue = '无'
def printList(item, keys):
    print(sep.join(list(map(lambda key:defaultValue if item[key] == '' else str(item[key]), keys.split(',')))))
head = {'pos':'位置','pro':'职业','dis1':'词缀1','dis2':'词缀2','minLevel':'最低','rank':'优先','canUpdate':'能升级','names':'干员'}
headNames = 'pos,pro,dis1,dis2,minLevel,rank,canUpdate,names'
printList(head, headNames)
for item in openedl:
    item['names'] = ','.join(list(map( lambda s:s['name'], item['result'])))
    printList(item, headNames)
```
